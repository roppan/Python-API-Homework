```python
news = ["@BBC", "@CBS", "@CNN", "@Fox", "@NYTimes"]
counter = 1
sentiments = []
```

```python
for source in news:
    tweets = api.user_timeline(source, count=100)
    
    tweetcount = 1
    
    for tweet in tweets:
        print("Tweet %s: %s" % (counter, tweet["text"]))
        compound = analyzer.polarity_scores(tweet["text"])["compound"]
        pos = analyzer.polarity_scores(tweet["text"])["pos"]
        neu = analyzer.polarity_scores(tweet["text"])["neu"]
        neg = analyzer.polarity_scores(tweet["text"])["neg"]
        tweets_ago = tweetcount
        sentiments.append({"Media Source": source,
                           "Text":tweet["text"],
                           "Date": tweet["created_at"],
                           "Compound": compound,
                           "Positive": pos,
                           "Neutral": neu,
                           "Negative": neg,
                           "Tweet Count": tweetcount})
        tweetcount +=1
        counter +=1
```

```python
sentiments_analysis = pd.DataFrame.from_dict(sentiments)
```

```python
sentiments_analysis_csv = sentiments_analysis[['Media Source', 'Date', 'Text', 'Compound', 'Positive', 'Neutral', 'Negative', 'Tweet Count']]
sentiments_analysis_csv.head()
```

```python
sentiments_analysis_csv.to_csv("News Mood.csv")
```

```python
news_colors = {"BBC": "lightblue",
               "CBS": "green",
               "CNN":"red",
               "Fox":"blue",
               "NYTimes": "gold"}
```

```python
plt.scatter(sentiments_analysis_csv["Tweet Count"], sentiments_analysis_csv["Compound"])
plt.xlabel("Tweets Ago")
plt.ylabel("Tweet Polarity")
plt.title("Sentiment Analysis of Media Tweets (%s)" % (time.strftime("%m/%d/%Y")))
plt.xlim(100,-2, -1)
plt.ylim(-1,1)
for newsoutlet in news_colors.keys():
    df = sentiments_analysis_csv[sentiments_analysis_csv['Media Source'] == newsoutlet]
    sentiments_analysis = plt.scatter(df["Tweet Count"],df["Compound"], label = newsoutlet, color = news_colors[newsoutlet], edgecolor = "black", s=125)
plt.legend(bbox_to_anchor = (1,1), title = 'Media Sources')    
plt.show()
sentiments_analysis.figure.savefig('Sentiment Analysis of Media Tweets.png')
```
